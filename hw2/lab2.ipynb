```python
import numpy as np
import matplotlib.pyplot as plt

x = [-1.0, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0]
y = [-4.467, -0.452, 0.551, 0.048, -0.447, 0.549, 4.552]

plt.scatter(x, y)
plt.show()
```

```python
f = np.poly1d(np.array([2.004357, -0.954464]).astype(float))
print(f)
x0 = np.linspace(-2, 2, 1000)
y0 = f(x0)
plt.scatter(x, y)
plt.plot(x0, y0)
plt.show()
y_f = f(x)
print(y - y_f)
```
